# gradio 範例

```python
def myadd(n1,n2):
  return n1+n2, n1-n2, n1*n2, n1/n2
```

```python
import gradio as gr # 載入gradio套件

iface = gr.Interface(
    fn=myadd, # 處理函式名稱
    inputs=['number','number'], # 輸入資料格式
    outputs=['number','number','number','number'] # 輸出資料格式
)

iface.launch(share=True, debug=True) # 啟動gradio介面, 啟動外部連結
```

# 使用 stable diffusion pipeline 以及 自定義函數 透過 gradio介面操作 stable diffusion pipeline

```python
from diffusers import StableDiffusionPipeline # 載入stable diffusion pipeline
import torch

model_id = "SG161222/Realistic_Vision_V6.0_B1_noVAE" # 選擇SD模型
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")
```

```python
def gen_image(prompt):
  image = pipe(prompt).images[0]
  return image
```

```python
import gradio as gr

iface = gr.Interface(
    fn=gen_image,
    inputs='text',
    outputs='image'
)

iface.launch(share=True, debug=True)
```

# Stable Diffusion from gradio.Interface.from_pipeline

```python
from diffusers import StableDiffusionPipeline
import gradio as gr

pipe = StableDiffusionPipeline.from_pretrained("SG161222/Realistic_Vision_V6.0_B1_noVAE", allow_pickle=False)

gr.Interface.from_pipeline(pipe).launch()
```
