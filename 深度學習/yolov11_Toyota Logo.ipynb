偵測影片

```python
model_test = YOLO(r"C:\Users\TKU-STAFF\Downloads\Yolov11\runs\detect\train9\weights\best.pt")

# 即時處理影片
video_path = r"C:\Users\TKU-STAFF\Downloads\Yolov11\toyota.mp4"
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: 無法開啟影片。")
else:
    while True:
        ret, frame = cap.read()
        if not ret:
            print("影片結束或發生錯誤。")
            break

        # 使用模型進行物件偵測
        results = model_test.predict(source=frame, conf=0.8, save=False, save_txt=False)

        # 繪製偵測結果
        annotated_frame = results[0].plot()

        # 展示結果
        cv2.imshow("YOLO Detection", annotated_frame)

        # 按 'q' 鍵退出
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2


# 載入 YOLO 模型
model_test = YOLO(r"C:\Users\TKU-STAFF\Downloads\Yolov11\runs\detect\train9\weights\best.pt")

# 即時處理影片
video_path = r"C:\Users\TKU-STAFF\Downloads\Yolov11\toyota.mp4"
cap = cv2.VideoCapture(video_path)

# 確認影片是否成功開啟
if not cap.isOpened():
    print("Error: 無法開啟影片。")
else:
    # 取得影片的幀率和解析度
    fps = cap.get(cv2.CAP_PROP_FPS)
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

    # 定義輸出的影片檔案和格式
    output_path = r"C:\Users\TKU-STAFF\Downloads\Yolov11\output_video.mp4"
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # 設定為 .mp4 格式
    out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

    while True:
        ret, frame = cap.read()
        if not ret:
            print("影片結束或發生錯誤。")
            break

        # 使用模型進行物件偵測
        results = model_test.predict(source=frame, conf=0.8, save=False, save_txt=False)

        # 繪製偵測結果
        annotated_frame = results[0].plot()

        # 展示結果
        cv2.imshow("YOLO Detection", annotated_frame)

        # 儲存結果到新影片檔案
        out.write(annotated_frame)

        # 按 'q' 鍵退出
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    # 釋放資源
    cap.release()
    out.release()
    cv2.destroyAllWindows()
```
