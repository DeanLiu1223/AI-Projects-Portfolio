```python
from datasets import load_dataset
from transformers import AutoTokenizer
import pandas as pd
from datasets import Dataset
```

```python
df = pd.read_json("/content/output.jl", lines=True)
```

```python
df["content"]
```

```python
tokenizer = AutoTokenizer.from_pretrained('gpt2')
tokenizer.pad_token = tokenizer.eos_token
hf_dataset = Dataset.from_pandas(df)

def tokenize_function(examples):
    inputs = tokenizer(examples['content'], truncation=True, padding='max_length', max_length=128)
    inputs['labels'] = inputs['input_ids'].copy()
    return inputs

tokenized_datasets = hf_dataset.map(tokenize_function, batched=True)
```

```python
import torch
from transformers import AutoModelForCausalLM, Trainer, TrainingArguments

model = AutoModelForCausalLM.from_pretrained('gpt2')

tokenizer.pad_token = tokenizer.eos_token

training_args = TrainingArguments(
    output_dir='/content/gpt2tw',
    num_train_epochs=200,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    warmup_steps=500,
    weight_decay=0.01,
    report_to=None,
    fp16=True,
    eval_strategy="epoch"

)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets,
    eval_dataset=tokenized_datasets,
)
```

```python
import wandb
wandb.init(mode="disabled")
trainer.train()
```

```python
torch.save(model.state_dict(), './gpt2_model')
```

```python
import torch
import transformers
from transformers import AutoModelForCausalLM
from transformers import AutoTokenizer

# 選擇 GPU 或 CPU
device = 'cuda' if torch.cuda.is_available() else 'cpu'
# 準備 tokenizer 及 模型架構
tokenizer1 = AutoTokenizer.from_pretrained('gpt2')
tokenizer1.pad_token = tokenizer.eos_token

model1 = AutoModelForCausalLM.from_pretrained("gpt2", num_labels=2)

# 載入模型權重
model1.load_state_dict(torch.load('./gpt2_model'))
model1.eval()
# 將模型搬到 GPU/CPU
model1.to(device)
```

```python
keyword_pairs = [("行為", "習慣")]

for kw1, kw2 in keyword_pairs:
    prompt = f"根據『{kw1}』與『{kw2}』生成新聞標題或摘要："

    inputs = tokenizer1(prompt, return_tensors='pt').to(model1.device)

    outputs = model1.generate(
        **inputs,
        max_length=150,
        num_return_sequences=3,
        do_sample=True,
        top_k=50,
        top_p=0.95
    )

    print(f"\n 關鍵字配對：{kw1} + {kw2} \n")
    for i, output in enumerate(outputs):
        text = tokenizer1.decode(output, skip_special_tokens=True)
        print(f"  {i+1}. \n {text} \n \n")
```
